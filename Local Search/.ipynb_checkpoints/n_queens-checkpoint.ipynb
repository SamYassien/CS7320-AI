# Solving the n-Queens Problem using Local Search

Points: 10

## The n-Queens Problem

__Goal:__ Find an arrangement of $n$ queens on a $n \times n$ chess board so that no queen is on the same row, column or diagonal as any other queen.

__State space:__ An arrangement of the queens on the board. We restrict the state space to arrangements where there is only a single queen per column. We represent a state as an integer vector of length $n$, each number representing for one column (from left to write) the row the queen is located in. We will call a state a "board."

__Objective function:__ The number of pairwise conflicts (i.e., two queens in the same row/column/diagonal).

The optimization problem is to find the optimal arrangement $a^*$ of $n$ queens on the board:

$$a^* = \mathrm{argmin}_a [ \mathrm{conflicts}(a) ]$$

$$\text{s.t.}\ a \ \text{contains only one queen per column}$$

Note that for this problem there is always an arrangement $a^*$ with $\mathrm{conflicts}(a) = 0$. 

__Local move:__ Move one queen to a different position in its column.

## Helper functions


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import math


def random_board(n):
    """Creates a random board of size n x n. Note that only a single queen is placed in each column!"""
    
    return(np.random.randint(0,n-1, size = n))


def conflicts(board):
    """Caclulate the number of conflicts, i.e., the objective function."""
    
    board = np.array(board)
    
    n = len(board)
    conflicts = 0

    # check horizontal (we do not check vertical since the state space is restricted to one queen per col)
    for i in range(n): conflicts += math.comb(np.sum(board == i), 2)
    #print(f"Horizontal conflicts: {conflicts}")
    
    # check for each queen diagonally up and down (only to the right side of the queen)
    for j in range(n):
        q_up = board[j]
        q_down = board[j]
     
        for jj in range(j+1, n):
            q_up -= 1
            q_down += 1
            if board[jj] == q_up: conflicts += 1
            if board[jj] == q_down: conflicts += 1
        #print(f"Conflicts after queen {j}: {conflicts}")
       
    return(conflicts)
        

def show_board(board, cols = ['white', 'gray']):  
    """display the board"""
    
    n = len(board)
    
    # create chess board display
    display = np.zeros([n,n])
    for i in range(n):
        for j in range(n):
            if (((i+j) % 2) != 0): 
                display[i,j] = 1
    
    cmap = colors.ListedColormap(cols)
    fig, ax = plt.subplots()
    ax.imshow(display, cmap = cmap, 
              norm = colors.BoundaryNorm(range(len(cols)+1), cmap.N))
    ax.set_xticks([])
    ax.set_yticks([])
    
    # place queens. Note: 1 and j are switched. Unicode u265B is a black queen
    for j in range(n):
        plt.text(j, board[j], u"\u265B", fontsize = 48, 
                 horizontalalignment = 'center',
                 verticalalignment = 'center')
    
    print(f"Board with {conflicts(board)} conflicts.")
    plt.show()
```

## Create a board

```python
board = random_board(4)

show_board(board)
print(f"Queens (left to right) are at rows: {board}")
print(f"Number of conflicts: {conflicts(board)}")
```

## Steepest-ascend Hill Climbing Search [3 Points]

Calculate the objective function for all local moves (move each queen within its column) and always choose the best among all local moves.

```python
def  getNeighbour(board):
    min_conflicts = conflicts(board)
    neighbour = None
    for row in range (len(board)):
        for col in range (len(board)):
            if col != board[row]:
                copy = list(board)
                copy[row] = col
                if (conflicts(copy) <min_conflicts): #better state
                    min_conflicts = conflicts(copy)
                    neighbour = copy
    return neighbour
    
def steepest_hill_climbing(board, verb= False):

    current = board
    if verb: show_board(current)
    while True:
        #get the best available neighbour
        neighbour = getNeighbour(current)
        if not neighbour:
            break

        if conflicts(neighbour) >= conflicts(current): #best neighbour is not helpful
            break
        current = neighbour
        if verb: show_board(current)
        
    if verb: print("Reached optimum at ", conflicts(current), " conflicts\n")
    return current
```

```python
board = random_board(4)
board = steepest_hill_climbing(board, verb = True)
```

Steepest-ascend Hill Clibming Search doesn't gurantee a global optimum as it can get caught in a local optimum.

## Steepest-ascend Hill Climbing Search with Random Restarts [1 Point]

Steepest-ascend hill climbing will often end up in local optima. Restart the algorithm up to 100 times with a random board to find a better (hopefully optimal) solution.   

```python
def random_restart(board=None,iterations=100, verb= False):
    goal_conflicts = 0
    
    for count in range (iterations):
        board = random_board(4)
        if verb: print("iter ", count, " conflicts ", conflicts(board))
        board = steepest_hill_climbing(board)
        if ( conflicts(board) == goal_conflicts ): break
    
    return board
```

```python
random_restart(verb= True)
```

Using random starts with the Steepest-ascend Hill Clibming Search doesn't gurantee a global optimum, but it has a way better chance of stumbling upon the answer as it restarts repeatedly

## Stochastic Hill Climbing [1 Point]

Chooses randomly from among all uphill moves.

```python
import random

def  getNeighbours(board):
    min_conflicts = conflicts(board)
    neighbours = []
    for row in range (len(board)):
        for col in range (len(board)):
            if col != board[row]:
                copy = list(board)
                copy[row] = col
                if (conflicts(copy) < min_conflicts): #list of all uphill moves
                    neighbours.append(list(copy))
    return neighbours
    
def stochastic_hill_climbing(board, verb= False):

    current = board
    if verb: show_board(current)
    while True:
        # get list of all uphill moves
        neighbours = getNeighbours(current)
        if not neighbours:
            break
        # choose a random uphill move
        idx = np.random.randint(len(neighbours))
        neighbour = neighbours[idx]

        # can't find a better solution
        if conflicts(neighbour) >= conflicts(current): 
            break
        current = neighbour
        if verb: show_board(current)
        
    if verb: print("Reached optimum at ", conflicts(current), " conflicts\n")
    return current
```

```python
board = random_board(4)
board = stochastic_hill_climbing(board, verb= True)
```

Stochastic hill climbing randomly chooses randomly from uphill moves. It does not gurantee a global optimum but it could arrive at it unless it gets stuck where no there are no improving neighbors.

## First-choice Hill Climbing [1 Point]

First-choice hill climbing is a type of stochastic hill climbing that generates one random local neighbor at a time and accept it if it has a better objective function value than the current state.

```python
def  getNeighbour(board):
    min_conflicts = conflicts(board)
    max_randomization = len(board) * len(board)
    neighbour = None
    count = 0
    while count < max_randomization:
        # generate random neighbour 
        row = np.random.randint(len(board))
        col = np.random.randint(len(board))
        count += 1
        if col != board[row]:
            copy = list(board)
            copy[row] = col
            if (conflicts(copy) <min_conflicts): #if has a better objective function, return it
                neighbour = copy
                return neighbour
    
def firstChoice_hill_climbing(board, verb= False):

    current = board
    if verb: show_board(current)
    while True:
        # get the first generated neighbour with a better objective funtion
        neighbour = getNeighbour(current)
        if not neighbour:
            break
        # can't find a better solution
        if conflicts(neighbour) >= conflicts(current):
            break
        current = neighbour
        if verb: show_board(current)
        
    if verb: print("Reached optimum at ", conflicts(current), " conflicts\n")
    return current
```

```python
board = random_board(4)
board = firstChoice_hill_climbing(board, verb = True)
```

First-choice hill climbing randomly selects the first neighbor with a better objective function. This strategy is good when the number of available neighbors is large. However, it's easy for it to get stuck in a local optimim and not escape.

## Simulated Annealing [2 Points]

You have to experiment with the annealing schedule.

```python
def schedule(t, k=4, alpha=0.999):
    return (k * alpha ** t)

def  getNeighbour(board):
    neighbour = list(board)
    row = np.random.randint(len(board))
    col = np.random.randint(len(board))
    neighbour[row] = col
    return neighbour

def simulated_annealing(board, verb= False):
    current = board
    current_conflicts = conflicts(current)
    
    for t in range(9999999):
        T = schedule(t) #get temperature
        if T == 0 or conflicts(current) == 0:
            return current
        
        neighbour = getNeighbour(current)        
        if not neighbour:
            return current
        
        deltaE = conflicts(neighbour) - conflicts(current)

        if deltaE < 0 or np.random.rand() < math.exp(-deltaE/T): #if conflicts decrease or matches probability, accept neighbour         
            current = neighbour
            if verb:
                print ( "t = ", t )
                show_board(current)
    
    if verb: show_board(current)
    return current
```

```python
board = random_board(4)
board = simulated_annealing(board, verb = True)
```

The Simulated Annealing algorithm improves over the other algorithms but it comes on the expense of time. The Simulated Annealing gurantees a global maximum If temperature decreases with a slow enough rate. This could be a bit unrealistic with big problems, but it seems to reach the global maximum in my trails on sizes 4 to 9.

## Compare Performance [2 Points]

Use runtime, scalability, and best objective function value to compare the algorithms on boards of different sizes.  

For timing you can use the `time` package.

```python
import time
from tabulate import tabulate

algorithms = [steepest_hill_climbing, random_restart, stochastic_hill_climbing, firstChoice_hill_climbing , simulated_annealing]
algos_names = ["steepest", "rand_restart", 
               "stochastic", "1stChoice" , "sim_annealing"]

table = dict()
headers = ["Algorithms","Average time","result_conflicts"]

for n in range(4,9):
    for i in range(len(algorithms)):
        avg_time = 0
        for itr in range(20):
            
            board = random_board(n)
            t0 = time.time()
            result = algorithms[i](board)
            t1 = time.time()
            result_conflicts = conflicts(result)
            avg_time += (t1-t0) * 1e3;
        avg_time = avg_time / 20
        if i == 0 :
            table[n] = ( [[algos_names[i], avg_time, result_conflicts]] )
        else:
            table[n].append( [algos_names[i], avg_time, result_conflicts] )
 
print("Data collected.")
for n in range(4,9):
    print(n, "-queens puzzle: \n")
    print(tabulate(table[n], headers = headers) + "\n")
    print('-----------')
    time = [table[n][i][1] for i in range(len(algorithms))]
    numconflicts = [table[n][i][2] for i in range(len(algorithms))]

    y = [table[n][i][0] for i in range(len(algorithms))]
    plt.bar(y, time)
    plt.xlabel('Algorithms')
    plt.ylabel('Time (mlsec)')
    plt.title('Time graph for n = {}' .format(n))
    plt.show()
    
    y = [table[n][i][0] for i in range(len(algorithms))]
    plt.bar(y, numconflicts)
    plt.xlabel('Algorithms')
    plt.ylabel('Conflicts')
    plt.title('Num of conflicts graph for n = {}' .format(n))
    plt.show()


 
```
